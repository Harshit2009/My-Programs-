```python
#Summary
#Packet Sniffer - also known as a packet analyzer, protocol analyzer or network analyzer â€” is a piece of hardware or 
#software used to monitor network traffic. Sniffers work by examining streams of data packets that flow between computers
#on a network as well as between networked computers and the larger Internet.

#Project On packeT Sniffer

from tkinter import *  #for making of gui
from tkinter.scrolledtext import ScrolledText  #for using scroll bar
from scapy.all import *  #for intoducing packets
logging.basicConfig(filename='tmp.log',format='%(levelname)s %(asctime)s :: %(message)s',level=logging.DEBUG)
import sys

win = Tk()
win.title("Packet Sniffer")  #title of graphical user interface
win.geometry("1100x800")
win.configure(bg = "GREY")  #background color of window

txt1 = StringVar()  #text variables for entry boxes and lables
txt2 = StringVar()
txt3 = StringVar()
txt4 = StringVar()
txt5 = StringVar()
txt6 = StringVar()
txt7 = StringVar()

e1 = Entry(win, width = 80, borderwidth = 5, textvariable = txt1, bg = "white")   #for making entry box
e1.grid(row = 0, column = 1, columnspan = 3 , padx = 100, pady = 10)

e2 = Entry(win, width = 80, borderwidth = 5, textvariable = txt2, bg = "white")   
e2.grid(row = 2, column = 1, columnspan = 3, padx = 100, pady = 10)

e3 = Entry(win, width = 80, borderwidth = 5, textvariable = txt3, bg = "white")   
e3.grid(row = 3, column = 1, columnspan = 3, padx = 100, pady = 10)

e4 = Entry(win, width = 80, borderwidth = 5, textvariable = txt4, bg = "white")   
e4.grid(row = 4, column = 1, columnspan = 3, padx = 100, pady = 10)

e5 = Entry(win, width = 80, borderwidth = 5, textvariable = txt5, bg = "white")   
e5.grid(row = 5, column = 1, columnspan = 3, padx = 100, pady = 10)

e6 = Entry(win, width = 80, borderwidth = 5, textvariable = txt6, bg = "white")   
e6.grid(row = 6, column = 1, columnspan = 3, padx = 100, pady = 10)

t1 = ScrolledText(win, width = 50, height = 3, bg = "white")  #for mking text boxes along with scrolled text
t1.grid(row = 1, column = 1, columnspan = 3, padx = 100, pady = 10)

t2 = ScrolledText(win, width = 50, height = 3, bg = "white")
t2.grid(row = 7, column = 1, columnspan = 3, padx = 100, pady = 10)

t3 = ScrolledText(win, width = 50, height = 3, bg = "white")
t3.grid(row = 10, column = 1, columnspan = 3, padx = 100, pady = 10)

t4 = ScrolledText(win, width = 90, height = 3, bg = "white")
t4.grid(row = 9, column = 0, columnspan = 3, padx = 100, pady = 10)

t5 = ScrolledText(win, width = 90, height = 5, bg = "white")
t5.grid(row = 101, column = 0, columnspan = 3, padx = 100, pady = 10)

l1 = Label(win, text = 'NETWORK ADDRESS', font = ("Bradley Hand ITC",16, "bold"),
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l1.grid(row = 0, column = 0, columnspan = 1)  #for making labels

l2 = Label(win, text = 'RAW PACKET DATA', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l2.grid(row = 1, column = 0, columnspan = 1)

l3 = Label(win, text = 'SOURCE ADDRESS', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l3.grid(row = 2, column = 0, columnspan = 1)

l4 = Label(win, text = 'DESTINATION ADDRESS', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l4.grid(row = 3, column = 0, columnspan = 1)

l5 = Label(win, text = 'SOURCE PORT', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l5.grid(row = 4, column = 0, columnspan = 1)

l6 = Label(win, text = 'DESTINATION PORT', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l6.grid(row =5, column = 0, columnspan = 1)

l7 = Label(win, text = 'VERSION', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l7.grid(row = 6, column = 0, columnspan = 1)

l8 = Label(win, text = 'PACKET DATA', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, anchor = W, bg = "WHITE", fg = "BLACK")
l8.grid(row = 7, column = 0, columnspan = 1)

l9 = Label(win, text = 'NEW PACKET GENERTAED', font = ("Bradley Hand ITC",16, "bold"), 
           justify = LEFT, bg = "WHITE", fg = "BLACK")
l9.grid(row = 10, column = 0, columnspan = 1)

def choice1():   #function for radio button
    logging.debug("Choices")
    p = sniff(filter="tcp", count = 1)
    if(txt7.get() == "BINARY"):
        t4.insert(END, p[0].payload)
    if(txt7.get() == "UNICODE"):
        t4.insert(END, export_object(p[0].payload))
        
r1 = Radiobutton(win, text = "BINARY", variable = txt7, value = "BINARY", command = choice1, 
                 font = ("Bradley Hand ITC",17, "bold"), bg = "WHITE", fg = "BLACK")   #radio buttons
r1.grid(row = 8, column = 0)  

r2 = Radiobutton(win, text = "UNICODE", variable = txt7, value = "UNICODE", command = choice1, 
                 font = ("Bradley Hand ITC",17, "bold"), bg = "WHITE", fg = "BLACK") 
r2.grid(row = 8, column = 2)

txt7.set("BINARY")

def read():   #function for log file
    with open('tmp.log') as f:
        content = f.readlines()
        t5.insert(END, content)
        
def bytes():  #function for bytes count
    z = 0
    y = 0
    p = sniff(filter="tcp", count = 1)
    z = z + sys.getsizeof(p[0].summary())
    y = y + sys.getsizeof(p[0].payload)
    t5.insert(END, z)
    t5.insert(END, y)

def generate_packets(): #function for packet generation
    logging.debug("Packet Generation")
    p = TCP()
    t3.insert(END,p)
    return

def start_sniff():  #function for packet sniffing
    logging.debug("Sniff")
    p = sniff(filter="tcp", count = 1)
    q = raw(p[0])
    d1 = p[0].src
    d2 = p[0].dst
    d3 = p[0].sport
    d4 = p[0].dport
    d5 = p[0].version
    txt1.set(p[0].summary())
    txt2.set(d1)
    txt3.set(d2)
    txt4.set(d3)
    txt5.set(d4)
    txt6.set(d5)
    t1.insert(END,q)
    t2.insert(END,p[0].payload)
    return

generate_packets = Button(win, text = "GENERATE PACKETS", padx = 60, pady = 20, command = generate_packets, 
                          font = ("Bradley Hand ITC",16, "bold"), bg = "WHITE", fg = "BLACK")  #making of buttons

start_sniff = Button(win, text = "START SNIFFING", padx = 60, pady = 20, command = start_sniff, 
                     font = ("Bradley Hand ITC",16, "bold"), bg = "WHITE", fg = "BLACK")

logs = Button(win, text = "LOGS", padx = 60, pady = 20, command = read, 
              font = ("Bradley Hand ITC",16, "bold"), bg = "WHITE", fg = "BLACK")

bytes = Button(win, text = "BYTES", padx = 30, pady = 20, command = bytes, 
               font = ("Bradley Hand ITC",16, "bold"), bg = "WHITE", fg = "BLACK")

generate_packets.grid(row =100, column = 0)
start_sniff.grid(row = 100, column = 2, columnspan = 1)
logs.grid(row = 100, column = 1, columnspan = 1)
bytes.grid(row = 100, column = 3)

win.mainloop()  #for holding the ouput screen
```
